## Ex3-001
1. Скачать данные по ссылке https://www.kaggle.com/datasets/ionaskel/laptop-prices.  
2. Считать данные с помощью pandas.
3. Вывести на экран первые 5 строк.

```python
import pandas as pd
```

```python
file_data = 'laptop_price.csv'
```

```python
df = pd.read_csv(file_data, encoding='ISO-8859-1')
```

```python
df.head()
```

### Ex3-001-1
Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU.

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split()[0])
```

```python
df.head()
```

### Ex3-001-2
Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя.

```python
def convert_TB_to_GB(string: str) -> int:
    new_string = string.split()[0]
    if new_string.endswith('TB'):
        result = float(new_string.replace('TB', ''))*1024
    else:
        result = new_string.replace('GB', '')
    return int(result)
```

```python
df['Memory_amount'] = df['Memory'].apply(convert_TB_to_GB)
```

### Ex3-001-3
Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.).

```python
df['Memory_type'] = df['Memory'].apply(lambda x: x.split()[1])
```

```python
df.head()
```

### Ex3-001-4
Удалите признаки Memory и ScreenResolution.

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
```

```python
df.head()
```

## Ex3-004
Составьте несколько сводных таблиц.

### Ex3-004-1
Найдите среднюю стоимость ноутбуков в зависимости от компании производителя.  
Отсортируйте от меньшей стоимости к большей.

```python
df.groupby('Company').agg({'Price_euros': 'mean'}).sort_values('Price_euros')
```

### Ex3-004-2
Найдите минимальную, среднюю и максимальную стоимости ноутбуков в зависимости от производителя процессора.

```python
df.groupby('Cpu_Company').agg({'Price_euros': ['mean', 'max', 'min']})
```

### Ex3-004-3
Постройте таблицу с подсчетом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ.

```python
df.pivot_table(index='Cpu_Company',
               columns='Ram',
               values='Weight',
               aggfunc='count',
               fill_value='0'
               )
```

### Ex3-004-4
Постройте таблицу с подсчетом средней стоимости ноутбуков в данных в зависимости от операционной системы и GB памяти.

```python
df.pivot_table(index='OpSys',
               columns='Memory_amount',
               values='Price_euros',
               aggfunc='mean',
               fill_value=0
               )
```

```python
df.groupby(['OpSys', 'Memory_amount']).\
agg({'Price_euros': 'mean'}).\
unstack('Memory_amount').\
fillna(0)
```
