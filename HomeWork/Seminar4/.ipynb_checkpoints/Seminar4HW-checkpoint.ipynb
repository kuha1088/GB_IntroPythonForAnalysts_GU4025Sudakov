### Task 1.4
Изучите распределение года постройки:

## Task 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.  
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
data = df[['price', 'sqft_living', 'floors', 'condition', 'grade',
           'waterfront', 'view', 'yr_built', 'sqft_lot', 'bedrooms',
           'bathrooms']]
```

```python
corr_matrix = data.corr()
```

```python
corr_matrix = np.round(corr_matrix, 4)
```

```python
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
```

```python
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True, linewidths = 0.5, cmap='coolwarm');
```

Значимое влияние на стоимость оказывают жилая площадь, оценка, количество ванных комнат и в меньшей степени вид из окна и количество спален. При этом, близость к набережной влияет на вид из окна, но не влияет на стоимость прямо (что странно).
Логически обусловлен отрицательный коэффициент корреляции для состояния дома и года постройки (чем старщше дом - тем хуже состояние).

```python
plt.boxplot(df['sqft_living']);
```

Жилая площадь - самый значимый показатель для стоимости.
Основной диапазон площадей объектов сгруппирован около уровня 2 000 sqft. Возможно, типовые дома.
